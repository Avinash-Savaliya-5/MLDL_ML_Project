### 1. Import Pandas, Numpy, Matplotlib Libraries & Load datasets...

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Done...")
df = pd.read_csv('cardio_train.csv', sep=';')
print("Done...")
print("Done...")
```

### 2. Explore columns & their datatypes

```python
print(df.info())
print(df.dtypes)
```

### 3. Print first 5 rows & datatypes of all columns

```python
print(df.head(5))
print(df.dtypes)
```

### 4. Find missing values

```python
df.isnull().sum()
```

### 5. Remove ID column

```python
df = df.drop("id", axis=1)
```

```python
df
```

### 6. Handeling duplicates rows

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df.shape
```

### 7. Convert age into years

```python
df['age'] = df['age'].astype(int)
df
```

### 8. Correlation Matrix

```python
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
```

### 9. Remove unrelated columns(gender & height)

```python
df = df.drop('gender', axis=1)
```

```python
df
```

### 10. Drop rows where age is not between 18-70

```python
df = df.drop(df[(df['age']<18) | (df['age']>70)].index)
```

### 11. Find mean & median of age

```python
print(f"Average of age : {df['age'].mean().astype(int)}")
print(f"Median of age  : {df['age'].median().astype(int)}")
```

### 12. Create Age_Category column for categories of age

```python
print(f"Minimum age : {(df['age']).min()}")
print(f"Maximum age : {(df['age']).max()}")
```

```python
df['age_category'] = np.where((df['age'] >= 25) & (df['age'] <= 30), 'Young',
                     np.where((df['age'] >= 31) & (df['age'] <= 50), 'Middle_age',
                     np.where((df['age'] >= 51) & (df['age'] <= 70), 'Senior', '' )))
```

```python
df
```

### 13. Count age_catogory wise total records

```python
df['age_category'].value_counts()
```

### 14. Plot a bar age_category wise safe records

```python
age_counts = df[df['cardio'] == 0]['age_category'].value_counts()
plt.bar(age_counts.index, age_counts.values)
plt.show()
```

### 17. Total safe & risky

### 18. Plot a Histogram of cholesterol column for find frequency of each unique values

```python
df['cholesterol'].hist(bins = 3)
plt.xlabel('Age')
plt.ylabel('frequency')

plt.show()
```

### 19. Pie Chart of Cholesterol Categories

```python
chol_counts = df['cholesterol'].value_counts()

plt.pie(chol_counts, labels=chol_counts.index, autopct='%1.1f%%')
plt.title('Cholesterol Distribution')
plt.show()
```
